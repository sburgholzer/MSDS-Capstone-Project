```python
import os
import re
import pandas as pd
files = [os.path.join(path, name) for path, subdirs, files in os.walk(r"D:\MSDS-Capstone-Project\implementation\textfiles") for name in files]
```

```python
files12Z = []
for myFile in files:
    if re.match(r".*[A-Z]{3}-[0-9]{2}-[0-9]{4}-12Z.txt", myFile):
        files12Z.append(myFile)
    
```

```python
files00Z = []
for myFile in files:
    if re.match(r".*[A-Z]{3}-[0-9]{2}-[0-9]{4}-00Z.txt", myFile):
        files00Z.append(myFile)
```

```python
data = pd.read_csv(files12Z[0], sep=",")
data = data[['VAR', 'VAL']].T
data.columns = data.iloc[0]
data = data.drop('VAR')
myDate =files12Z[0].split("\\")[-1].split(".txt")[0].split("-")
myDate = myDate[0] + "-" + myDate[1] + "-" + myDate[2]
myDate = [myDate]
data['Date'] = myDate
data.set_index("Date", inplace = True) 

df12Z = data
df12Z
```

```python
for idx, myFile in enumerate(files12Z):
    if myFile != files12Z[0]:
        data = pd.read_csv(files12Z[idx], sep=",")
        data = data[['VAR', 'VAL']].T
        data.columns = data.iloc[0]
        data = data.drop('VAR')
        myDate =files12Z[idx].split("\\")[-1].split(".txt")[0].split("-")
        myDate = myDate[0] + "-" + myDate[1] + "-" + myDate[2]
        myDate = [myDate]
        data['Date'] = myDate
        data.set_index("Date", inplace=True)
        df12Z = df12Z.append(data) 
```
